$$f(w) = 14(w-2)^2$$
$$w_n = w_{(n-1)} - lr \cdot \nabla _wf(w_{(n-1)})$$

```python
def gd(w = 10, lr = 0.02, itera_times = 20):
    results = [w]
    for i in range(itera_times):
        w = w - lr*28*(w - 2)
        results.append(w)
    return results
```

```python
res = gd()
```

```python
def show_trace(res):
    f_line = np.arange(-6, 10, .1)
    plt.plot(f_line, [14 * np.power(x-2, 2) for x in f_line])
    plt.plot(res, [14 * np.power(x-2, 2) for x in res], '-o')
    plt.xlabel('x')
    plt.ylabel('Loss(x)')
```

```python
show_trace(res)
```
